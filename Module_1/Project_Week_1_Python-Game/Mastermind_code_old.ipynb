# Mastermind code


```python
# Begin

import random
import re

# Welcome message AM-11

print("Hey welcome to IronMastermind !!")

# Randomise color combination AM-12

Color_dictionary = {"g": "Green", "b": "Blue", "w": "White", "k": "Black", "y": "Yellow", "r": "Red", "p": "Pink"}

G_combination = []

Color_list = list(Color_dictionary.keys())

G_combination = random.choices(Color_list, k = 4) 

#print(G_combination)

#Variable Initialisation AM-13

answer = (0,0)
count_turns = 0
G_user_combination = []

# Loop start

while answer != (4,0):

    # Ask the user for a combination AM-14

    #Loop untill input_valid

    question = """"Please enter a suite of 4 colors;

    Use ‘k' for Black ; ‘g’ for Green ; ‘b' for Blue ; ‘w’ for White ; ‘y’ for Yellow ; 'r’ for Red ; 'p’ for Pink

    For exemple : type gbwy and press enter for Green-Blue-White-Yellow"""
    
    #Test the validity of the input ,  G_user_combination
    
    input_valid = False
    
    while input_valid == False:
        MyStr = input(question)
        MyStr = str(MyStr)
        MyStr = MyStr.lower()
        input_valid = False
        pattern = '[kgbwyrp]{4}'
        if (re.findall(pattern, MyStr) != []):
            if len(MyStr)==4:
                input_valid = True
        if input_valid == False:
            print('Please enter a correct suite of 4 colors')
    
    G_user_combination=list(MyStr)
    count_turns += 1

    # Test the combination and build answer AM-15

    #Count the number of good placed

    L_good = 0

    for i in range(0,4):
        if G_combination[i] == G_user_combination[i]:
            L_good += 1

    #Count the number of misplaced elements

    L_miss_placed = 0

    for i in G_combination:
        if i in G_user_combination:
            L_miss_placed += 1
    L_miss_placed = L_miss_placed - L_good        

    #Build an answer 

    answer = (L_good, L_miss_placed)

    #Display answer AM-16

    #Build the answer message:

    if L_good != 0 or L_miss_placed != 0:
        print("You have", L_good, "elements at the right place")
        print("You have", L_miss_placed, "elements at the wrong place")
        print("You have",12 -  count_turns, "turns left")
    else:
        print("No elements are present")
        print("You have",12 - count_turns, "turns left")

    if count_turns == 12:
        break

if answer == (4,0):
    # Display win message AM-17
    print("Bravo ! you master my mind !!")
else:
    # Display lose message AM-18
    print("Oh no, no more try left! You lost")



```
