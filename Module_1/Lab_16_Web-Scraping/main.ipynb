#### Display the trending Python repositories in GitHub.

The steps to solve this problem is similar to the previous one except that you need to find out the repository names instead of developer names.

````python
# This is the url you will scrape in this exercise
url = 'https://github.com/trending/python?since=daily'
````

````python
response = requests.get(url).text

soup = BeautifulSoup(response)

soup.prettify()

articles = soup.find_all('article')

repo_names = [] 

for i in articles:
    repo_names.append(i.h2.text.replace('\n','').split('/')[1].strip())
    
print(repo_names)
    
````

#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page.

````python
response = requests.get(url).content
sp = BeautifulSoup(response, 'html.parser')

language = '#js-link-box-fr > strong'

language1 = 'strong'

len(sp.select(language1))

languages = []
for i in sp.select(language1):
    languages.append(i.get_text())

new_languages = []
new_languages = languages.pop()
new_languages = languages.pop(0)

article = '#js-link-box-fr > small > bdi'

article1 = 'small > bdi'

len(sp.select(article1))

articles = []
for i in sp.select(article1):
    articles.append(i.get_text())
    
wiki = pd.DataFrame({'Languages': languages, 'Related articles': articles})
wiki
````
